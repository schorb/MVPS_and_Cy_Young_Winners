```python
df_obp_avg = df.groupby(df.columns[0]).mean()[['OBP']]
df_slg_avg = df.groupby(df.columns[0]).mean()[['SLG']]
```

One shortfall of predicting using sabermetric data is the fact that + attributes cannot be calculated without the season being played.  I'll have to improvise a bit  

OPS+ should be an easy fix.  the stand-in will be 2018's OPS average.  

```python
fans_obp = df_obp_avg.loc[2018][0]
```

```python
fans_slg = df_slg_avg.loc[2018][0]
```

```python
FANS['OPS+'] = round(100*((FANS['OBP']/fans_obp)+(FANS['SLG']/fans_slg) - 1), 0).astype(int)
```

```python
FANS = FANS[FANS['PA'] >= 450]
```

```python
FANS.head()
```

Solved.  WRC will be a little harder to work around.  It's based off of wOBA which is in the projections (thank god)

'Baseball is senseless, meaningless, chaotic, it symbolizes nothing.  It's a mess.  But it's also beautiful'

-- Jon Bois

```python
FANS['wOBA'].mean()
```

```python
wOBA_scale = 1.226
```

```python
run_PA = .117
```

```python
wRC = (((FANS['wOBA']-FANS['wOBA'].mean())/wOBA_scale) + run_PA)*FANS['PA'] 
```

```python
FANS['wRC+'] = round(wRC/wRC.mean()*100, 0).astype(int)
```

```python
FANS['TTO%'] = round((FANS['BB'] + FANS['HBP'] + FANS['SO'] + FANS['HR'])/FANS['PA'], 3)
```

Last thing to reconcile is IBB%

```python
(df['IBB']/df['BB']).mean()
```

About 8% of all walks are intentional.  We'll use that to infer everyone's projected IBB

```python
FANS['IBB'] = round(FANS['BB']*.0835,0).astype(int)
```
