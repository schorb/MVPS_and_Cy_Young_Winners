Eric Gagne, Dodgers Closer, won the 2003 NL Cy Young award.  However, at less than 162 IP, he did not qualify for pitching statistics and is not included in this list

```python
cya = [('Jacob deGrom', 2018),('Blake Snell', 2018),('Max Scherzer', 2017),('Corey Kluber', 2017),('Max Scherzer', 2016),('Rick Porcello', 2016),('Jake Arrieta', 2015),('Dallas Keuchel', 2015),('Clayton Kershaw', 2014),('Corey Kluber', 2014),('Clayton Kershaw', 2013),('Max Scherzer', 2013),('R.A. Dickey', 2012),('David Price', 2012),('Clayton Kershaw', 2011),('Justin Verlander', 2011),('Roy Halladay', 2010),('Felix Hernandez', 2010),('Tim Lincecum', 2009),('Zack Greinke', 2009),('Tim Lincecum', 2008),('Cliff Lee', 2008),('Jake Peavy', 2007),('CC Sabathia', 2007),('Brandon Webb', 2006),('Johan Santana', 2006),('Chris Carpenter', 2005),('Bartolo Colon', 2005),('Roger Clemens', 2004),('Johan Santana', 2004),('Roy Halladay', 2003),('Randy Johnson', 2002),('Barry Zito', 2002),('Randy Johnson', 2001),('Roger Clemens', 2001),('Randy Johnson', 2000),('Pedro Martinez', 2000),('Randy Johnson', 1999),('Pedro Martinez', 1999),('Tom Glavine', 1998),('Roger Clemens', 1998),('Pedro Martinez', 1997),('Roger Clemens', 1997),('John Smoltz', 1996),('Pat Hentgen', 1996),('Greg Maddux', 1995),('Randy Johnson', 1995)]
```

```python
df['CYA'] = 0
```

Same general function as before

```python
for v in range(len(df)):
    #separate the cya winners
    for (i, j) in cya:
    
        #match the winners with their respective seasons
        if (i, j) == (df['Name'][v], df['Season'][v]):
            #profit
            df['CYA'][v] = 1
        else:
           pass
```
